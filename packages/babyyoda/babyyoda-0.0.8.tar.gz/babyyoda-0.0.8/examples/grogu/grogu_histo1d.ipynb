# Grogu Hist1D

```python
import babyyoda.grogu as yoda
from babyyoda import loc, project, rebin
```

## Create Histo1D

```python
# Simple 2D histogram, 1 entry per bin, some overflows that should not matter
h = yoda.Histo1D(10, 10, 20, title="test")
for i in range(-10, 22):
    for j in range(-10, 32):
        h.fill(i, (i - j) ** 2)
h
```

## Access

```python
h[0]
```

```python
h[loc(11)]
```

## Setting

```python
h[0] = h[-1]
h
```

## Slicing

```python
h[4:6]
```

## Rebin

```python
h[loc(14) : loc(19) : rebin(2)]
```

## Project/Sum axis

```python
h[::project].plot()
h[loc(10) : loc(12) : project].plot()
```

## String

```python
print(h.to_string())
```

```python
print(h[:2].to_grogu_v2().to_string())
```
