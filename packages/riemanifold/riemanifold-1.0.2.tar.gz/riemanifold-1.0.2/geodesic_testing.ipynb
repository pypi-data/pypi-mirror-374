```python
import torch
import numpy as np
from riemanifold import *
```

```python
sphere = UVSphere(r=np.float64(1.0))
```

```python
curve = coord_lerp([-np.pi/2, 0.1], [0.0, np.pi/2])
```

```python
sphere.show_curve(curve)
```

```python
ts = torch.linspace(0.0, 1.0, 100)
```

```python
def v(t): # input: single number
    t = torch.as_tensor(t)
    return torch.autograd.functional.jacobian(curve, t, create_graph=True)
def dv_dt(t): # input: single number
    t = torch.as_tensor(t)
    return torch.autograd.functional.jacobian(v, t)
```

```python
t = 0.5
```

```python
G_t = sphere.Gamma(curve(torch.as_tensor(t)))
torch.einsum('kij, i, j -> k', G_t, v(t), v(t))
```
